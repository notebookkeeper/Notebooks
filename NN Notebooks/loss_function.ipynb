```python
import tensorflow as tf
import numpy as np
```

```python
# Binary Cross Entropy
# class = yes , class = no
y_true = [[0, 1], [0, 0], [1,0]] ## data
y_pred = [[-18.6, 0.51], [2.94, -12.8],[12,9]] ## last output layer
bce = tf.keras.losses.BinaryCrossentropy(from_logits=True)
bce(y_true,y_pred).numpy()
```

```python
# Binary Cross Entropy
# class = yes , class = no
y_true = [[0, 1], [0, 0], [1,0]] ## data
y_pred = [[-18.6, 0.51], [2.94, -12.8],[12,9]] ## last output layer
bce = tf.keras.losses.BinaryCrossentropy()
bce(y_true,y_pred).numpy()
```

```python
y_true = [[0, 1], [0, 0], [1,0]]
y_pred = [[0, 1], [1, 0],[0.9526,0.0474]]
mse = tf.keras.losses.MeanSquaredError()
print(mse(y_true, y_pred).numpy())

mae = tf.keras.losses.MeanAbsoluteError()
print(mae(y_true, y_pred).numpy())
```

```python
# Binary Cross Entropy
y_true = [[0, 1], [0, 0], [1,0]]
y_pred_linear_operation = [[-18.6, 0.51], [2.94, -12.8],[12,9]]
y_pred_proba = tf.keras.activations.softmax(tf.constant(y_pred_linear_operation))
bce = tf.keras.losses.BinaryCrossentropy()
bce(y_true,y_pred_proba).numpy()
```

```python
# Binary Cross Entropy
y_true = [[0, 1, 0], [0, 0, 1], [1,0, 0]]
y_pred = [[0.45, 0.91, 0.51], [2.94, 3.12, 12.8],[12, 9, 8.9]]
proba = tf.keras.activations.softmax(tf.constant(y_pred))
print(np.round(proba,3))
bce = tf.keras.losses.CategoricalCrossentropy()
bce(y_true,proba).numpy()
```

```python
# Binary Cross Entropy
y_true = [[0, 1, 0], [0, 0, 1], [1,0, 0]]
y_pred = [[0.45, 0.91, 0.51], [2.94, 3.12, 12.8],[12, 9, 8.9]]
proba = tf.keras.activations.softmax(tf.constant(y_pred))
print(np.round(proba,3))
cce = tf.keras.losses.CategoricalCrossentropy()
cce(y_true,proba).numpy()
```

```python
y_true = [1, 2]
y_pred = [[0.05, 0.95, 0, 0], [0.1, 0.8, 0.1, 0]]
scce = tf.keras.losses.SparseCategoricalCrossentropy()
proba = tf.keras.activations.softmax(tf.constant(y_pred))
print(np.round(proba,3))
scce(y_true, y_pred).numpy()
```
