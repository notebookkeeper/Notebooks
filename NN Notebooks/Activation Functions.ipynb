```python
import tensorflow as tf
from tensorflow.keras import activations
import numpy as np

# Refer to https://keras.io/api/layers/activations/
```

```python
x = tf.constant([-10, -5, 0.0, 5, 10], dtype = tf.float32)
x
```

```python
# Relu = max(x, 0)
# resut of the linear op of a layer = [-10.,  -5.,   0.,   5.,  10.]
tf.keras.activations.relu(x)
```

```python
# Sigmoid sigmoid(x) = 1 / (1 + exp(-x))
x = tf.constant([-1.5, -1, 0.5, 1, 2], dtype = tf.float32)
np.around(tf.keras.activations.sigmoid(x),4)
```

```python
# Softmax
x_reshaped = tf.constant([[-1.5, -1, 0.5, 1, 2]], dtype = tf.float32)
print(x_reshaped)
proba = tf.keras.activations.softmax(x_reshaped)
print('proba',np.around(proba,4))
print(tf.reduce_sum(proba) )
```
